```python
plt.imshow(im_1)
plt.show()
```

Resmi siyah beyaz yapmak için: her pixel'i RGB'nin ortalamasına eşitliyoruz.

```python
new_image=np.zeros((m,n),dtype=float)#im_1 resmi ile aynı boyuta resim oluşturduk
for i in range(m):
    for j in range(n):
        s=(im_1[i,j,0]+im_1[i,j,1]+im_1[i,j,2])/3#rgb ortalamasını aldık
        new_image[i,j]=s
plt.imshow(new_image,cmap='gray')
plt.show()
plt.imsave('siyahbeyazcanakkale.jpg',new_image,cmap='gray')
```

DOKSAN DERECE DÖNDÜRME

```python
new_image=np.zeros((n,m),dtype=float)
for i in range(m):
    for j in range(n):
        s=(im_1[i,j,0]+im_1[i,j,1]+im_1[i,j,2])/3
        new_image[j,i]=s #i,j degerlerini değiştirdik 
plt.imshow(new_image,cmap='gray')
plt.show()
plt.imsave('doksanderececanakkale.jpg',new_image,cmap='gray')
```
